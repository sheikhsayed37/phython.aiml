```python
square = {x: x**2 for x in range(1,10) if x%2==0}
print(square) #modified version nice

#output:
#{2: 4, 4: 16, 6: 36, 8: 64}
```

```python
square = {}               # ১. খালি dictionary নেওয়া হলো

for x in range(1, 10):    # ২. 1 থেকে 9 পর্যন্ত loop
    if x % 2 == 0:        # ৩. check even number
        square[x] = x**2  # ৪. key = x, value = x²

print(square)
```
