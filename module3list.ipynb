## list comprehension basic


```python
random_list=[x for x in range(20)]
print(random_list)
```

```python
random_list=[x for x in range(1,20)]
print(random_list)
```

```python
random_list=[x*2 for x in range(1,20)]
print(random_list)
```

```python
random_list=[x for x in range(40) if x%2==0]
print(random_list)
```

```python
random_list=[x for x in range(40) if x%2!=0]
print(random_list)
```

```python
fruits=['apple','banana','orange','grape','kiwi']
capitalize_fruits=[]

for fruit in fruits:   # i in range() er moto kaj kore 
    capitalize_fruits.append(fruit.capitalize())

print(capitalize_fruits)    
```
