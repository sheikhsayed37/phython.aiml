```python
#list 
list1=[10,20,30,40,50,60,60,70,200]
print(list1)

```

```python
#accessing list items
list1=[10,20,30,40,50,60,60,70,200]
print(list1[0]) #first item
print(list1[3]) #fourth item
```

```python
#adding items to the list

list1=[10,20,30,40,50,60,60,70,200]
list1.append(300) #add item at the end of the list
print(list1)
```

```python
# insert item at specific index
 
list1=[10,20,30,40,50,60,60,70,200]
list1.insert(3,588) 
print(list1)
```

```python
list1=[10,20,30,40,50,60,60,70,200]
list1.pop() #remove last item
print(list1)
```
